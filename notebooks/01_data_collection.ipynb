```python
import yfinance as yf
import pandas as pd
import os

ticker="RELIANCE.NS"

df= yf.download(ticker,start="2019-01-01", end="2024-01-01")
#yf itself return a pandas dataframe type by default

# print(df.columns)

df.columns = df.columns.get_level_values(0)



print(df.shape) # gives rows and coloumns of data
print(df.head(10)) #shows first 10 rows
print(df.tail(10)) #last 10 rows
print(df.dtypes) #shows datatypes of coloumn
print(df.isnull().sum()) #to check how many missing data
```

```python
os.makedirs("../data",exist_ok=True)
df.to_csv("../data/RELIANCE_raw.csv")
print("Saved successfully")


df_loaded=pd.read_csv("../data/RELIANCE_raw.csv",index_col=0, parse_dates=True)
print(df_loaded.head(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("../data/RELIANCE_raw.csv",index_col=0, parse_dates=True)

plt.figure(figsize=(14,5))
plt.plot(df["Close"])
plt.title("Reliance Industries - CLosing Price(2019-2024)")
plt.xlabel("Date")
plt.ylabel("Price (INR)")
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(14,4))
plt.bar(df.index,df["Volume"], color="steelblue", alpha=1) #format for this is plt.bar(xvalue,yvalue)
plt.title("Reliance - Daily Trading volume")
plt.xlabel("Data")
plt.ylabel("Volume")
plt.tight_layout()
plt.show()
```

```python
df["Daily_Return"] = df["Close"].pct_change()#curr-yester/yesterday

print(df["Daily_Return"].describe())#describe gives important stats about the Daily_Return column

plt.figure(figsize=(14,4))
plt.plot(df["Daily_Return"],alpha=0.6,color="darkgreen")
plt.axhline(0,color="red",linestyle="--",linewidth=1)
plt.title("Reliance - Daily Returns")
plt.xlabel("Data")
plt.ylabel("Daily Return")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(df["Daily_Return"].dropna(),bins=100,color="steelblue",edgecolor="black",alpha=0.7)
plt.title("Distribution of Daily Returns")
plt.xlabel("Daily Return")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
