```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("../data/RELIANCE_features_v2.csv", index_col=0, parse_dates=True)
print(df.shape)
print(df.isnull().sum())

df=df.dropna()
print(df.shape)
print(df.isnull().sum())

print(df.head(10))

```

```python
X = df.drop(columns=["Target"])
Y = df["Target"]

print(X.shape)
print(Y.shape)
print(Y.value_counts())
```

Why do we split this data? X is everything the model is allowed to see and Y is what we are trying to predict


```python
split_index=int(len(df)*0.8)
print("Toal rows: ", len(df))
print("Split index: ",split_index)
print("Training ends on: ",df.index[split_index])
print("Test starts on: ", df.index[split_index+1])

X_train=X.iloc[:split_index]
X_test=X.iloc[split_index:]
Y_train=Y.iloc[:split_index]
Y_test=Y.iloc[split_index:]

print("X_train shape: ", X_train.shape)
print("X_test shape: ", X_test.shape)

```

What date range is your training set? data range for training is from start till 2023-01-11
What date range is your test set? from 2023-01-12 till end
Why is it important that these two ranges don't overlap? no future bias for ML model 

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()

X_train_scaled=scaler.fit_transform(X_train) #fit_transform is a combination of fit() and transform(), fit() is used to make the system learn min and max of the 
                                            #min and max of the dataset, then trasnform() will make it in range 0 to 1
X_test_scaled=scaler.transform(X_test) #we dont use fit here as we dont want the data to know about the min amx of the test set
                                        #we want data to just use the fit made by training set and scale the dataset according to that only

X_train_scaled = pd.DataFrame(X_train_scaled,columns=X.columns, index=X_train.index) #no longer a data frame it is like a numpy array now so we tranform back
#doing this fit_transform changes the datatype from pandas dataframe to numpy array, we need to trasnform it back

X_test_scaled = pd.DataFrame(X_test_scaled, columns=X.columns, index=X_test.index)


print(X_train_scaled.head(5))
print(X_train_scaled.describe().round(3))
```

After prinitng we see 25% 50% 75%, what are those???? bascialy they tell that 50% values are below this number, 25% values are below this number 
(also called quartiles)



After scaling, what should the min and max of every column be? 0 and 1

Why do we fit on training data only and not the full dataset? we do not want the ML model to have a dataset which is scaled based on the test data, so we only scale based on training data and use that same scale and apply it to test also

What would go wrong if we scaled the full dataset first and then split? still the same issue, the scaling would be dependant on the later 20% of the data as well which we dont want. 

```python
X_train_scaled.to_csv("../data/X_train.csv")
X_test_scaled.to_csv("../data/X_test.csv")
Y_train.to_csv("../data/Y_train.csv")
Y_test.to_csv("../data/Y_test.csv")

print("All files saved")
print(f"Training samples: {len(X_train_scaled)}")
print(f"Test samples: {len(X_test_scaled)}")
```
